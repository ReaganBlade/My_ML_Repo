# ML Model to Predict the Quality of Apples

### importing data set


```python
import numpy as np
import pandas as pd

df = pd.read_csv("D:\\Personal\\Data Science\\DataSets\\apple_quality.csv")
df
```

```python
df.dropna(inplace=True)
```

```python
print(df)
```

```python
df
```

```python
# transforming Data in Quality Column to 1s and 0s

df['Quality'] = df['Quality'].map({'good':1, 'bad':0})
```

```python
df
```

```python
# data preparation
y = df['Quality']
```

```python
y
```

```python
X = df.drop('Quality', axis=1)
X
```

```python
# Data Splitting

from sklearn.model_selection import train_test_split

X_train, X_test, y_test, y_train = train_test_split(X, y, test_size=0.5, random_state=100)
```

```python
# print(X_train)
print(y_train)
```

```python
X_test
```

```python
# Model Building Using Linear Regression
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# Applying the model to make a prediction
lrTrainPredict = lr.predict(X_train)
lrTestPredict = lr.predict(X_test)
```

```python
print(lrTrainPredict)
```

```python
print(lrTestPredict)
```

```python
# Evaluate Model Performance
```

```python
from sklearn.metrics import mean_squared_error as mse, r2_score as r2s

lrTrainMse = mse(y_train, lrTrainPredict)
lrTrainR2 = r2s(y_train, lrTrainPredict)

lrTestMse = mse(y_test, lrTestPredict)
lrTestR2 = r2s(y_test, lrTestPredict)
```

```python
print('LR MSE (Train): ', lrTrainMse)
print('LR R2 (Train): ', lrTrainR2)
print('LR MSE (Test): ',lrTestMse)
print('LR R2 (Test): ', lrTestR2)
```

```python
lrResults = pd.DataFrame(["LinearRegression", lrTrainMse, lrTrainR2, lrTestMse, lrTestR2]).transpose()
lrResults.columns = ['Method', 'Training MSE', "Training R2", "Testing MSE", "Testing r2"]
```

```python
lrResults
```

```python
# Random Forest
# Training Model 

from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(X_train, y_train)
```

```python
# applying the model to make a prediction

rfTrainPred = rf.predict(X_train)
rfTestPred = rf.predict(X_test)
```

```python
# Evaluate Model Performance

from sklearn.metrics import mean_squared_error, r2_score

rfTrainMse = mean_squared_error(y_train, rfTrainPred)
rfTrainR2 = r2_score(y_train, rfTrainPred)

rfTestMse = mean_squared_error(y_test, rfTestPred)
rfTestR2 = r2_score(y_test, rfTestPred)
```

```python
rfResults = pd.DataFrame(['Random Forest', rfTrainMse, rfTrainR2, rfTestMse, rfTestR2]).transpose()
rfResults.columns = ['Method', 'Training MSE', "Training R2", "Testing MSE", "Testing r2"]

rfResults
```

```python
# Model Comparison

df_models = pd.concat([lrResults, rfResults], axis = 0)

df_models.reset_index(drop = True)
```
